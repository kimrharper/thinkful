```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import calendar
from matplotlib.pyplot import subplots, show
%matplotlib inline
```

Q1:

```python
binormdist = np.random.binomial(100, 0.5, 1000)
plt.plot(binormdist)
plt.show()
```

```python
plt.hist(binormdist)

# Add a vertical line at the mean.
plt.axvline(binormdist.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(binormdist.mean() + binormdist.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(binormdist.mean()-binormdist.std(), color='y', linestyle='dashed', linewidth=2) 

plt.show()
print("MEAN: {}".format(binormdist.mean()))
print("STAND DEV: {}".format(binormdist.std()))
```

The highest point in the graph matches the mean so it appears the finding the mean is relevant to the data set. The inner quartiles also appear to include a majority of the data so it is possible to limit the data to the inner quartile ranges if needed.

```python
rand1 = np.random.normal(5, .5, 100)
rand2 = np.random.normal(10, 1, 100)
```

```python
plt.hist(rand1)
plt.show()

plt.hist(rand2)
plt.show()

rand3 = rand1 + rand2
```

```python
plt.hist(rand3)

# Add a vertical line at the mean.
plt.axvline(rand3.mean(), color='y', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(rand3.mean() + rand3.std(), color='y', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(rand3.mean()-rand3.std(), color='y', linestyle='dashed', linewidth=2) 

plt.show()
print("MEAN: {}".format(rand3.mean()))
print("STAND DEV: {}".format(rand3.std()))
```

Data appears to be left skewed. Doesn't appear to be normal as the mean value is less than the values on both sides of it.
