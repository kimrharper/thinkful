```python
import pandas as pd
import matplotlib as plt
import numpy as np
import csv
import seaborn as sns
```

```python
with open('../data/cleveland.data',encoding = "ISO-8859-1") as input_file:
   lines = input_file.readlines()
   newLines = []
   for line in lines:
      newLine = line.strip().split()
      newLines.append(newLine)

heartdata = [] 
while len(newLines) > 0:
    heartrow = [] 
    for sublist in newLines[0:10]:
        for item in sublist:
            heartrow.append(item)  
        del newLines[0:10]
    heartrow.append('cleveland')
    heartdata.append(heartrow)
```

```python
colnames=['id','ccf','age','sex','painloc','painexer','relrest','pncaden','cp','trestbps','htn','chol','smoke',
              'cigs','years','fbs','dm','famhist','restecg','ekgmo','ekgday','ekgyr','dig','prop','nitr','pro',
              'diuretic','proto','thaldur', 'thaltime','met','thalach','thalrest','tpeakbps','tpeakbpd','dummy',
              'trestbpd', 'exang','xhypo','oldpeak', 'slope', 'rldv5','rldv5e','ca','restckm','exerckm','restef',
              'restwm', 'exeref','exerwm','thal','thalsev','thalpul','earlobe','cmo', 'cday', 'cyr','num','lmt',
              'ladprox','laddist','diag','cxmain','ramus','om1','om2','rcaprox','rcadist','lvx1','lvx2','lvx3','lvx4',
              'lvf','cathef','junk','name']
```

```python
dfHeart = pd.DataFrame(data=heartdata,columns=colnames)
```
